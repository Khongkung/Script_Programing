```python
# Snippet A
score =85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
else:
    grade = "C"
print(f"Grade: {grade}") # What's the output?
# ผลลัพธ์ที่ได้คือ Grade: B

# Block B: Vectorized Filter
phrase ="Python 2026 Engine"
consonants = {c.lower() for c in phrase if c.isalpha() and c.lower() not in 'aeiou'}
print(f"Unique consonants: {consonants}")
#ผลลัพธ์ที่ได้คือ Unique consonants: p,y,t,h,n,g เพราะคำสั่งจะคักเอาแค่ตัวอักษรและไม่เอาสระ aeiou

# Snippet B
temp = 25
is_raining = True
if temp > 20 and not is_raining:
    print("Perfect day for a picnic!")
elif temp <= 20 or is_raining:
    print("Stay indoors or bring an umbrella.")
else:
    print("Unsure about the weather.")
# What's the output?
#ผลลัพธ์ที่ได้คือ Stay indoors or bring an umbrella. เพราะ or คือถ้าใช่อย่างใดอย่างนึงในคำสั่งก็จะแสดงผลได้

# Snippet C
age =17
has_license = True
if age >= 18:
    if has_license:
        print("You can drive legally.")
    else:
        print("You need a license to drive.")
else:
    print("You are too young to drive.")
# What's the output?
#ผลลัพธ์ที่ได้คือ You are too young to drive. เพราะ age < 18
```

==START GAME=   
เกมนี้เป็นเกมเนื้อเรื่องตอบ ใช่/ไม่ใช่ แบบข้อความ

เริ่มเกมแล้วจะมีคำบรรยายฉาก
คุณตอบคำถามด้วย
y หรือ yes = ใช่
n หรือ no = ไม่ใช่
แต่ละคำตอบจะพาคุณไปยังทางเลือกถัดไป
ไม่มีคะแนน ไม่มีแพ้ชนะอย่างเดียว
ทุกเส้นทางจะจบด้วยเรื่องราวดี ๆ และจบสวย

```python
print("\n=== เกมผจญภัยแห่งป่าเวทมนตร์ ===\n")
print("คืนนี้ดวงจันทร์ส่องสว่างเหนือป่าสีเงิน คุณเดินทางมาถึงทางแยกหนึ่ง...")
print("ตอบด้วย 'y' ถ้าใช่ หรือ 'n' ถ้าไม่ใช่\n")

choice1 = input("คุณจะเดินตามแสงที่เปล่งประกายจากต้นไม้ใหญ่ไหม? (y/n): ").strip().lower()

if choice1 == 'y' or choice1 == 'yes':
    print("\nคุณก้าวเข้าไปในแสง ใจคุณทั้งตื่นเต้นและสงบ")
    print("แสงขอบฟ้าพาคุณผ่านป่าสีเงินจนคุณมาถึงลานกลมกลางป่า")
    print("ที่นั่นมีสุนัขจิ้งจอกพูดได้ และลมพัดกลิ่นหอมของดอกไม้ที่ไม่เคยเห็น")
    choice2 = input("มันขอให้คุณช่วยค้นหากุญแจวิเศษใต้ก้อนหินโบราณไหม? (y/n): ").strip().lower()

    if choice2 == 'y' or choice2 == 'yes':
        print("\nคุณยืนค้นหาด้วยความระมัดระวัง และพบกุญแจทองคำซ่อนใต้ใบไม้")
        print("จิ้งจอกยิ้มอย่างขอบคุณ แล้วนำคุณไปยังประตูที่ซ่อนอยู่ในรากไม้ใหญ่")
        choice3 = input("คุณจะใช้กุญแจไขประตูนั้นเพื่อดูว่ามีอะไรอยู่ข้างในไหม? (y/n): ").strip().lower()

        if choice3 == 'y' or choice3 == 'yes':
            print("\nประตูเปิดออกสู่สวนลับที่เต็มไปด้วยแสงระยิบระยับ")
            print("นางฟ้าป่าเวทมนตร์รอคุณอยู่ และมอบคำอธิษฐานให้หนึ่งข้อ")
            print("คุณขอให้ป่าวิเศษคุ้มครองเพื่อนใหม่และให้คุณพบความสุขที่แท้จริง")
            print("จบบริบูรณ์: คุณกลับออกมาพร้อมหัวใจที่เปลี่ยนไป และเรื่องราวในป่าจะกลายเป็นนิทานที่เล่าต่อ")
        elif choice3 == 'n' or choice3 == 'no':
            print("\nคุณตัดสินใจทิ้งประตูไว้และเดินกลับผ่านป่าด้วยความสงบ")
            print("เสียงจิ้งจอกยังคงตามคุณไป และคุณรู้สึกว่าบางครั้งความลับก็สวยงามเมื่อไม่ถูกเปิดเผย")
            print("จบบริบูรณ์: คุณออกจากป่าพร้อมความทรงจำลึกลับและความสุขที่อ่อนโยน")
        else:
            print("\nคุณหยุดคิดอยู่ครู่หนึ่ง และป่าก็เงียบลงเหมือนรอคำตอบ")
            print("ในที่สุดคุณเลือกเดินออกจากลานอย่างช้าๆ โดยไม่เร่งรีบ")
            print("จบบริบูรณ์: ความเป็นไปได้ยังคงอยู่ แต่คุณกลับบ้านด้วยความสงบและความตื่นเต้นในใจ")

    elif choice2 == 'n' or choice2 == 'no':
        print("\nคุณปฏิเสธอย่างสุภาพ และจิ้งจอกปรายตามองคุณด้วยดวงตาที่เข้าใจ")
        print("มันชี้ไปทางน้ำตกแห่งความสงบซึ่งส่องประกายใต้แสงจันทร์")
        choice3 = input("คุณจะเดินทางไปน้ำตกเพื่อฟังเสียงธรรมชาติและหาคำตอบใจไหม? (y/n): ").strip().lower()

        if choice3 == 'y' or choice3 == 'yes':
            print("\nที่น้ำตก คุณนั่งลง ฟังเสียงน้ำไหล และพบว่าทุกอย่างในชีวิตมีการไหลของมันเอง")
            print("คืนหนึ่งในป่าเปลี่ยนเป็นบทเรียนที่สวยงาม และคุณเรียนรู้ว่าการเลือกของเราไม่จำเป็นต้องถูกหรือผิด")
            print("จบบริบูรณ์: คุณกลับออกมาพร้อมใจที่อ่อนโยนและความสงบที่ไม่เคยรู้จักมาก่อน")
        elif choice3 == 'n' or choice3 == 'no':
            print("\nคุณเลือกเดินต่อไปบนเส้นทางของตัวเอง โดยไม่จำเป็นต้องได้คำตอบในทันที")
            print("แสงจันทร์ยังคงส่องนำทาง และคุณรู้ว่าการเดินทางก็เป็นสิ่งที่มีค่าเท่ากับจุดหมาย")
            print("จบบริบูรณ์: คุณออกจากป่าด้วยรอยยิ้มและความมั่นใจในทางที่เลือก")
        else:
            print("\nคำตอบไม่ชัดเจน แต่ป่าก็ยังคงยอมรับคุณ")
            print("คุณนั่งมองดอกไม้ไฟฟ้าที่ส่องสว่างตรงหน้า และรู้สึกว่าแม้ไม่เลือก คำตอบยังอยู่ในหัวใจ")
            print("จบบริบูรณ์: คุณกลับบ้านพร้อมความสงบและความหวัง")

    else:
        print("\nคำตอบไม่ชัดเจน แต่ป่าก็ยังคงอ่อนโยน")
        print("คุณเดินค้นหาตัวเองท่ามกลางแสงและเงา และพบว่าการไม่เร่งรีบคือการผจญภัยรูปแบบหนึ่ง")
        print("จบบริบูรณ์: คุณออกจากป่าด้วยความรู้สึกอบอุ่น และความทรงจำที่สวยงาม")

elif choice1 == 'n' or choice1 == 'no':
    print("\nคุณเลือกเดินออกจากแสง เข้าไปยังทางเดินเลียบแม่น้ำ")
    print("สายลมเย็นพัดผ่าน เสียงน้ำดังเบาๆ เหมือนเพลงกล่อม")
    choice2 = input("ขณะที่เดิน คุณเห็นเด็กน้อยร้องไห้ คุณจะเข้าไปช่วยเขาไหม? (y/n): ").strip().lower()

    if choice2 == 'y' or choice2 == 'yes':
        print("\nคุณเดินเข้าไปหาเด็กน้อยและช่วยให้เขากลับบ้านได้อย่างปลอดภัย")
        print("ในทางกลับ คุณได้พบกองทัพนกที่โบยบินเป็นรูปหัวใจ")
        print("เด็กน้อยบอกว่าความเมตตาของคุณทำให้ครอบครัวเขามีความหวัง")
        choice3 = input("คุณจะร่วมฉลองกับชาวบ้านที่งานเลี้ยงแสงจันทร์ไหม? (y/n): ").strip().lower()

        if choice3 == 'y' or choice3 == 'yes':
            print("\nงานเลี้ยงเต็มไปด้วยเสียงหัวเราะและแสงระยิบระยับ")
            print("คุณได้รับผลึกสีฟ้าที่ส่องแสงเมื่อมีคนพูดความจริง")
            print("จบบริบูรณ์: คุณกลับบ้านพร้อมของขวัญและความทรงจำที่ทำให้หัวใจพองโต")
        elif choice3 == 'n' or choice3 == 'no':
            print("\nคุณเลือกเดินกลับในความเงียบ แต่หัวใจอิ่มเอมไปด้วยการช่วยเหลือ")
            print("คุณรู้ว่าสิ่งที่ดีที่สุดคือการทำดีโดยไม่ต้องเป็นผู้ชม")
            print("จบบริบูรณ์: คุณกลับบ้านพร้อมความสุขที่เรียบง่ายและจริงใจ")
        else:
            print("\nคำตอบไม่ชัดเจน แต่ผู้คนยังยิ้มให้คุณ")
            print("ค่ำคืนนั้นเต็มไปด้วยความอบอุ่น และคุณก็รู้สึกว่าทุกการกระทำมีความหมาย")
            print("จบบริบูรณ์: คุณได้ความหวังกลับไปด้วยหัวใจที่อ่อนโยน")

    elif choice2 == 'n' or choice2 == 'no':
        print("\nคุณเลือกพักผ่อนใต้ต้นไม้ใหญ่ริมแม่น้ำ")
        print("เสียงน้ำไหลและแสงดาวทำให้คุณสงบ จนพบว่าแต่ละช่วงเวลาก็เป็นของขวัญ")
        print("เวลาผ่านไปช้า คุณได้เขียนความในใจลงบนใบไม้แห้ง และปล่อยให้มันลอยไปกับสายน้ำ")
        print("จบบริบูรณ์: คุณได้เรียนรู้ว่าการให้เวลากับตัวเองก็เป็นชัยชนะชนิดหนึ่ง")

    else:
        print("\nคำตอบไม่ชัดเจน แต่แม่น้ำยังคงไหลอย่างสวยงาม")
        print("คุณนั่งฟังเสียงธรรมชาติแล้วรู้สึกว่าทุกอย่างอยู่ในจังหวะที่ถูกต้อง")
        print("ในท้ายที่สุด คุณตระหนักว่าการเป็นตัวเองในค่ำคืนนั้นคือเรื่องที่ดีที่สุด")
        print("จบบริบูรณ์: คุณกลับบ้านพร้อมความสงบและความหวัง")

else:
    print("\nคำตอบนั้นไม่ชัดเจน แต่ป่าก็ยังยินดีต้อนรับคุณ")
    print("คุณเดินเล่นใต้แสงดาว และรู้ว่าทุกการตัดสินใจคือการผจญภัย")
    print("คุณได้เจอคนแปลกหน้าที่เดินตามทางกลับบ้านเดียวกัน")
    print("แม้จะไม่มีคำตอบชัดเจน กลับเป็นค่ำคืนที่เต็มไปด้วยความอบอุ่นและเรื่องเล่าที่คุณจะไม่ลืม")
    print("จบบริบูรณ์: คุณได้รับความรู้สึกอบอุ่นและมิตรภาพจากธรรมชาติ")
```
