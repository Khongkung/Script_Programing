เกมทายตัวเลขระบบจะสุ่มตัวเลข 1-100 มา 1 ตัวเลขและให้เราทายจนกว่าจะถูก
ถ้าคำตอบมากกว่าตัวเลขที่สุ่ม print("น้อยกว่านี้") ถ้าคำตอบน้อยกว่าตัวเลขที่สุ่ม print("มากกว่านี้") ถ้าคำตอบเท่ากับตัวเลขที่สุ่ม print("ยินดีด้วยคุณทายถูก เลขที่ใช่คือ{number}")


```python
import random

number = random.randint(1, 100)

while True:
    input_number = int(input("Guess a number between 1 and 100: "))
    if input_number < number:
        print("Too low! Try again.")
    if input_number > number:
        print("Too high! Try again.")
    if input_number == number:
        print("The number was " + str(number))
        print("Congratulations! You guessed the correct number.")
        break
```
