```python
my_integer = 10
my_float = 3.14
my_string = "Hello, Python!"
my_boolean = True

print("--- Variable Display ---")
print(f"Integer: {my_integer} (Type: {type(my_integer).__name__})")
print(f"Float:   {my_float} (Type: {type(my_float).__name__})")
print(f"String:  '{my_string}' (Type: {type(my_string).__name__})")
print(f"Boolean: {my_boolean} (Type: {type(my_boolean).__name__})")
print("------------------------")
```

```python
# ประกาศตัวแปรสำหรับคำนวณ
num1 = 25
num2 = 7
num_float = 12.5

print("\n=== การคำนวณพื้นฐาน ===")
print(f"ตัวเลขที่ 1: {num1}")
print(f"ตัวเลขที่ 2: {num2}")
print(f"ตัวเลขทศนิยม: {num_float}")
print("-----------------------")

# การคำนวณ
addition_result = num1 + num2
subtraction_result = num1 - num2
multiplication_result = num1 * num2
division_result = num1 / num2
exponent_result = num1 ** 2
compound_calculation = (num1 + num2) * num_float

# แสดงผลลัพธ์
print(f"ผลรวม ({num1} + {num2}): {addition_result}")
print(f"ผลต่าง ({num1} - {num2}): {subtraction_result}")
print(f"ผลคูณ ({num1} * {num2}): {multiplication_result}")
print(f"ผลหาร ({num1} / {num2}): {division_result:.2f} (ปัดเศษ 2 ตำแหน่ง)")
print(f"ยกกำลังสอง ({num1} ** 2): {exponent_result}")
print(f"การคำนวณแบบผสมผสาน (({num1} + {num2}) * {num_float}): {compound_calculation}")
print("======================\n")
```

```python
# กำหนดค่าความกว้างและความยาว
width = 10
length = 25

# คำนวณพื้นที่
area = width * length

# แสดงผลลัพธ์
print(f"--- การคำนวณพื้นที่สี่เหลี่ยมผืนผ้า ---")
print(f"ความกว้าง: {width} หน่วย")
print(f"ความยาว: {length} หน่วย")
print(f"พื้นที่สี่เหลี่ยมผืนผ้า: {area} ตารางหน่วย")
print("-------------------------------------")
```

```python
# Define the temperature in Celsius
celsius = 30

# Calculate the temperature in Fahrenheit
fahrenheit = (celsius * 1.8) + 32

# Display the result
print(f"--- Temperature Conversion ---")
print(f"Celsius temperature: {celsius}°C")
print(f"Fahrenheit temperature: {fahrenheit}°F")
print("------------------------------")
```
